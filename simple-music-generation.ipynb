```python
import pretty_midi as pm
```

```python
new_melody = pm.PrettyMIDI()
```

```python
# add instrument
new_instrument = pm.Instrument(program=49)
```

```python
scale_notes = ['G4', 'A4', 'B4', 'C4', 'D4', 'E4', 'F#4']
```

```python
note_lengths = [1.0, 1.0, 0.5, 1.0, 1.0, 1.0, 0.5]
```

```python
import random

melody_length = 32
end_time = 0

for k in range(melody_length):

    # start time of new note is end time of previous note
    start_time = end_time

    # generate random note
    random_note = random.choice(scale_notes)
    note_number = pm.note_name_to_number(random_note)

    # set note duration
    #note_duration = note_lengths[scale_notes.index(random_note)]
    note_duration = note_lengths[k % 7]

    end_time = start_time + note_duration

    # create and add new note
    new_note = pm.Note(pitch=note_number, velocity=100, start=start_time, end=end_time)
    new_instrument.notes.append(new_note)
```

```python
new_melody.instruments.append(new_instrument)
new_melody.write('output.mid')
```
