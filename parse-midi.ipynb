```python
import pretty_midi as pm
```

```python
# load midi file
midi_file = "sample.mid"
midi_data = pm.PrettyMIDI(midi_file)
```

```python
# total length of audio
duration = midi_data.get_end_time()
print(duration)
```

```python
# extract notes
from pretty_midi.utilities import note_number_to_name

for instrument in midi_data.instruments:
  if not instrument.is_drum:
    print(f"{instrument.name} | {instrument.program}")

  for note in instrument.notes:
    note_name = note_number_to_name(note.pitch)
    note_duration = note.end - note.start
    print(f"{note_name} | {note.velocity} | {note_duration:.2f}")
```
